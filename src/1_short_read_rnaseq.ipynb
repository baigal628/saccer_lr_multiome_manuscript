```r
library("DESeq2")
library("ggplot2")
library("ggrepel")
library("ggcorrplot")
library("dplyr")
library(stringr)
library(purrr)
library("tibble")
library(dplyr)
library(tidyr)
library(ComplexHeatmap)
library("pals")
library(ggpubr)
library(tximport)
library(DESeq2)
library("apeglm")
theme_set(
    theme_classic(base_size = 12)
)
```

```r
rsem_out = '/private/groups/brookslab/gabai/projects/yeastMeth/data/rna/totalrna/rsem/'
```

### isoform level counts

```r
outdir = '../data/1_total_rna_rsem'
```

```r
samples <- gsub(".isoforms.results", "", list.files(rsem_out, ".isoforms.results$"))
files <- file.path(rsem_out, list.files(rsem_out, ".isoforms.results$"))
names(files) <- samples
txi <- tximport(files, dropInfReps = TRUE, type="rsem", txIn = TRUE, txOut = TRUE, abundanceCol="TPM")
saveRDS(txi, file = '../data/1_total_rna_rsem_iso_txi.obj')
samples
write.table(txi$counts[,c(samples[c(4:6)], samples[c(1:3)])],  paste0(outdir, "_iso_counts.tsv") ,sep = "\t", quote = FALSE, col.names = TRUE, row.names = TRUE)
write.table(txi$abundance[,c(samples[c(4:6)], samples[c(1:3)])],  paste0(outdir, "_iso_TPM.tsv") ,sep = "\t", quote = FALSE, col.names = TRUE, row.names = TRUE)
```
